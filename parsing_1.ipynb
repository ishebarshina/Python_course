По этому фрейму вычислите среднее значение вызовов пожарных машин в месяц в одном округе Москвы, округлив до целых

Примечание: найдите среднее значение вызовов, без учета года

Cначала отредактируем данные

```python
data_1.AdmArea.value_counts().sort_index()
```

```python
data_1.AdmArea.replace('Центральный административный округ ', 'Центральный административный округ', inplace=True)
data_1.AdmArea.replace('Восточный административный округ ', 'Восточный административный округ', inplace=True)
data_1.AdmArea.replace('Западный административный округ ', 'Западный административный округ', inplace=True)
data_1.AdmArea.replace(['Западный административный округ ', 'Запдный административный округ'], 'Западный административный округ', inplace=True)
data_1.AdmArea.replace('Зеленоградский административный округ ', 'Зеленоградский административный округ', inplace=True)
data_1.AdmArea.replace('Северо-Востосный административный округ ', 'Северо-Востосный административный округ', inplace=True)
data_1.AdmArea.replace(['Северо-Западный административный округ ', 'Северо-Закпадный административный округ', 'Северо-Заподный административный округ'], 'Северо-Западный административный округ', inplace=True)
data_1.AdmArea.replace(['Троицкий и Новомосковский  административные округа ', 'Троицкий и Новомосковский  административные округа', 'Троицкий и Новомосковский административные округ а', 'Троицкий и Новомосковский административные округа', 'Троицкий и Новомосковский административный округа ', 'Троицкий и новомосковский административные округа', 'Троицкий и Новомосковский административные округа ', 'Троицкий и Новомосковский административный округа'], 'Троицкий и Новомосковский административные округа', inplace=True)
data_1.AdmArea.replace(['Юго=Восточный административный округ', 'Юго-Восточнный административный округ', 'Юго-Востосный административный округ'], 'Юго-Восточный административный округ', inplace=True)
data_1.AdmArea.replace('Северно-Западный административный округ', 'Северо-Западный административный округ', inplace=True)
data_1.AdmArea.replace(['Северо-Восточный административный округ ', 'Северо-Востосный административный округ'], 'Северо-Восточный административный округ', inplace=True)
data_1.AdmArea.replace('Южный административный округ ', 'Южный административный округ', inplace=True)
data_1.AdmArea.replace('Северный административный округ ', 'Северный административный округ', inplace=True)
data_1.AdmArea.replace('Юго-Восточный административный округ ', 'Юго-Восточный административный округ', inplace=True)
data_1.AdmArea.replace(['Юго-Западный административный округ ', 'Юго-западный административный округ', 'ЮгоЗападный административный округ'], 'Юго-Западный административный округ', inplace=True)
data_1.AdmArea.value_counts().sort_index()
```

Посмотрим по каждому месяцу в каждом округе:

```python
data_1.groupby(['AdmArea', 'Month']).Calls.mean().sort_index().apply(round)
```

Посмотрим отдельно для какого-нибудь одного округа:

```python
data_1[data_1['AdmArea']=='Восточный административный округ'].groupby('Month').Calls.mean().sort_index().apply(round)
```

**Найдем для каждого округа среднее значение вызовов в месяц**

```python
data_1.groupby('AdmArea').Calls.mean().sort_index().apply(round)
```

Какое среднее значение вызовов пожарных машин в месяц в одном округе Москвы в 2015-2019 годах с точностью до целых?

```python
data_1[data_1['Year'].isin(range(2015,2020))].groupby('AdmArea').Calls.mean().sort_index().apply(round)
```

```python
round(data_1[data_1['Year'].isin(range(2015,2020))].groupby('AdmArea').Calls.mean().mean())
```
