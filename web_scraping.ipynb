```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
url = 'https://www.kinopoisk.ru/top/'
```

```python
r = requests.get(url)
html = bs(r.content)
```

По тегу "а" = anchore (якорь) можно найти все ссылки в html тексте

```python
links = html.find_all("a")
links
```

```python
type(links)
```

 The HREF is an attribute of the anchor tag, which is also used to identify sections within a document.

```python
# будем искать ссылки на фильмы и на страницы
def find_links(links, keyword):
    links_filtered = {} # чтобы ссылки были уникальными
    for link in links:
        if link.has_attr("href") and link["href"].find(keyword) > -1:
            # атрибут href содержит ссылку на страницу фильма
            links_filtered[link["href"]] = 1
    return list(links_filtered.keys())
```

Найдем все фильмы на этой странице (там всего 5 страниц с фильмами, вроде)

```python
films = []
films.extend(find_links(html.find_all("a"), "/film/"))
```

```python
films
```

```python
pages = []
pages.extend(find_links(html.find_all("a"), "/?page"))
```

```python
pages
```

```python
for page in pages:
    r1 = requests.get("https://www.kinopoisk.ru" + page)
    html1 = bs(r1.content)
    films.extend(find_links(html1.find_all("a"), "/film/"))
```
