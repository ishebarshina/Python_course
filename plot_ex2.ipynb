```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
```

```python
data = pd.read_csv('https://video.ittensive.com/python-advanced/marathon-data.csv')
```

```python
def time_sec(x):
    l = x.split(':')
    return int(l[0])*3600 + int(l[1])*60 + int(l[2])
```

```python
data['split'] = data['split'].apply(lambda x: time_sec(x))
data['final'] = data['final'].apply(lambda x: time_sec(x))
```

```python
data
```

```python
sns.pairplot(data)
```

Видим, что коррелируют final и split

```python
scipy.stats.pearsonr(data['split'], data['final'])
```
